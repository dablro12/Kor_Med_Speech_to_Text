## Inference 

```python
from omnilingual_asr.models.inference.pipeline import ASRInferencePipeline

pipeline = ASRInferencePipeline(model_card="omniASR_CTC_7B")

audio_files = ["/workspace/kru_data/원천데이터/간호사/HA_0009/HA_0009-1-01-02-F-04-C.wav"]
lang = ["kor_Hang"]
transcriptions = pipeline.transcribe(audio_files, lang=lang, batch_size=len(audio_files)) 

print(f"정답 Transcription: 집안에 고혈압인 분 있나요?")
print(f"예측 Transcription: {transcriptions[0]}")
```
